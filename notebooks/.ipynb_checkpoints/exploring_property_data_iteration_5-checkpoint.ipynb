```python
import pandas as pd
import re
from datetime import datetime

# Read the sample data into a DataFrame
from io import StringIO


# Define the function to extract asking price at sale, sold price, sold date, and first list date
def get_price_details(price_changes: str):
    """
    Extracts the asking price at sale, sold price, sold date, and first list date from the Price Changes column.

    Args:
        price_changes (str): The value from the Price Changes column.

    Returns:
        tuple: A tuple containing:
            - Asking Price at Sale (float or None)
            - Sold Price (float or None)
            - Sold Date (str or None)
            - First List Date (str or None)
    """
    if not isinstance(price_changes, str):
        return (None, None, None, None)
    
    # Initialize variables
    sold_price = None
    sold_date = None
    asking_price_at_sale = None
    all_dates = []
    
    # Split the price changes by ';' to process each event separately
    entries = price_changes.split(';')
    
    for i, entry in enumerate(entries):
        entry = entry.strip()
        
        # Extract the date from the entry
        date_match = re.search(r"([A-Za-z]{3}\s+[A-Za-z]{3}\s+\d{1,2}\s+\d{4})", entry)
        if date_match:
            date_str = date_match.group(1)
            try:
                date_obj = datetime.strptime(date_str, "%a %b %d %Y")
                all_dates.append(date_obj)
            except ValueError:
                # Handle unexpected date formats
                pass
        
        # Match the 'Sold' event
        sold_match = re.match(
            r"Sold,\s*€([\d,]+),\s*([A-Za-z]{3}\s+[A-Za-z]{3}\s+\d{1,2}\s+\d{4})",
            entry,
            re.IGNORECASE
        )
        if sold_match:
            sold_price = float(sold_match.group(1).replace(',', ''))
            sold_date = sold_match.group(2)
            
            # Look for the next relevant event to find the asking price at sale
            for next_entry in entries[i+1:]:
                next_entry = next_entry.strip()
                asking_match = re.match(
                    r"(Sale Agreed|Price Drop|Created),\s*€([\d,]+),\s*[A-Za-z]{3}\s+[A-Za-z]{3}\s+\d{1,2}\s+\d{4}",
                    next_entry,
                    re.IGNORECASE
                )
                if asking_match:
                    asking_price_at_sale = float(asking_match.group(2).replace(',', ''))
                    break
            # Assuming only one 'Sold' event exists
            break
    
    # Determine the first list date (earliest date)
    if all_dates:
        first_list_date_obj = min(all_dates)
        first_list_date = first_list_date_obj.strftime("%a %b %d %Y")
    else:
        first_list_date = None
    
    return (asking_price_at_sale, sold_price, sold_date, first_list_date)

# Apply the function to extract the desired details and assign them to new columns
df[['Sold Asking Price', 'Sold Price', 'Sold Date', 'First List Date']] = df['Price Changes'].apply(get_price_details).apply(pd.Series)

# Convert the price columns to numeric types if they aren't already
df['Sold Asking Price'] = pd.to_numeric(df['Sold Asking Price'], errors='coerce')
df['Sold Price'] = pd.to_numeric(df['Sold Price'], errors='coerce')

# Convert 'First List Date' to datetime for better handling (optional)
df['First List Date'] = pd.to_datetime(df['First List Date'], format="%a %b %d %Y", errors='coerce')

# Display the updated DataFrame with the new columns
df[['Address', 'Price Changes', 'Sold Asking Price', 'Sold Price', 'Sold Date', 'First List Date']].head()
```

```python
import pandas as pd
import numpy as np
from sklearn.neighbors import NearestNeighbors

# --------------------------------------------
# 1. Data Preparation
# --------------------------------------------

# Assuming `house_rows` DataFrame is already available
# If not, you can load it using pd.read_csv or other relevant methods
# Example:
# house_rows = pd.read_csv('your_data.csv')

# Define the columns to use for model fitting
columns_subset_training = [
    'Address', 
    'Property Type', 
    'Beds', 
    'Baths', 
    'MyHome_Floor_Area_Value', 
    'Sold Asking Price', 
    'MyHome_Latitude', 
    'MyHome_Longitude'
]

# Subset the DataFrame
df_subset = house_rows[columns_subset_training].copy()

# Ensure 'Beds' and 'Baths' are treated as strings before extracting numeric values
df_subset['Beds'] = df_subset['Beds'].astype(str).str.extract('(\d+)').astype(float)
df_subset['Baths'] = df_subset['Baths'].astype(str).str.extract('(\d+)').astype(float)

# Convert other relevant columns to numeric, if not already
numeric_columns = ['MyHome_Floor_Area_Value', 'Sold Asking Price', 
                   'MyHome_Latitude', 'MyHome_Longitude']
for col in numeric_columns:
    df_subset[col] = pd.to_numeric(df_subset[col], errors='coerce')

# Drop rows with missing values in critical columns
df_subset.dropna(subset=['Beds', 'Baths', 'MyHome_Floor_Area_Value', 
                         'MyHome_Latitude', 'MyHome_Longitude'], inplace=True)

# Optional: Reset index after dropping rows
df_subset.reset_index(drop=True, inplace=True)

# --------------------------------------------
# 2. Define Helper Functions
# --------------------------------------------

def haversine_distance(lat1, lon1, lat2, lon2):
    """
    Calculate the Haversine distance between two points on the Earth.
    """
    R = 6371  # Earth radius in kilometers
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])  # Convert to radians
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat / 2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2)**2
    c = 2 * np.arcsin(np.sqrt(a))
    return R * c

def min_max_scale(feature):
    """
    Perform Min-Max scaling on a feature.
    """
    min_val = feature.min()
    max_val = feature.max()
    if max_val - min_val == 0:
        return feature - min_val  # Avoid division by zero
    return (feature - min_val) / (max_val - min_val)

def exponential_decay(distance, max_distance_km, decay_rate=1.0):
    """
    Apply exponential decay to the distance.
    
    Parameters:
    - distance: Distance in kilometers.
    - max_distance_km: Maximum distance considered.
    - decay_rate: Controls the rate of decay.
    
    Returns:
    - Decayed score between 0 and 1.
    """
    return np.exp(-decay_rate * (distance / max_distance_km))

def calculate_combined_score(df_subset, given_lat, given_lon, given_beds, 
                            given_baths, given_size, geo_weight=0.7, 
                            size_weight=60.0, beds_weight=20.0, 
                            baths_weight=20.0, max_distance_km=3, 
                            decay_rate=1.0):
    """
    Calculate a combined score based on geographic distance with exponential decline and feature similarity.
    
    Parameters:
    - df_subset: DataFrame with property data.
    - given_lat, given_lon: Latitude and Longitude of the target property.
    - given_beds, given_baths, given_size: Features of the target property.
    - geo_weight: Weight for geographic distance (0.7).
    - size_weight: Weight for the size feature (60).
    - beds_weight: Weight for beds feature (20).
    - baths_weight: Weight for baths feature (20).
    - max_distance_km: Maximum geographic distance to consider.
    - decay_rate: Rate of exponential decay for distance.
    
    Returns:
    - df_sorted: DataFrame sorted by the combined score.
    """
    # Calculate Haversine distance for each property
    df_subset['Geo Distance (km)'] = haversine_distance(
        given_lat, given_lon, 
        df_subset['MyHome_Latitude'], 
        df_subset['MyHome_Longitude']
    )
    
    # Filter properties within the specified geographic distance
    df_filtered = df_subset[df_subset['Geo Distance (km)'] <= max_distance_km].copy()
    
    if df_filtered.empty:
        print("No properties found within the specified geographic distance.")
        return df_filtered
    
    # Apply exponential decay to geographic distance
    df_filtered['Geo Score'] = exponential_decay(
        df_filtered['Geo Distance (km)'], 
        max_distance_km, 
        decay_rate
    )
    
    # Prepare KNN features
    knn_features = df_filtered[['MyHome_Floor_Area_Value', 'Beds', 'Baths']].copy()
    
    # Apply Min-Max scaling and weights
    knn_features_scaled = pd.DataFrame()
    knn_features_scaled['MyHome_Floor_Area_Value'] = min_max_scale(knn_features['MyHome_Floor_Area_Value']) * size_weight
    knn_features_scaled['Beds'] = min_max_scale(knn_features['Beds']) * beds_weight
    knn_features_scaled['Baths'] = min_max_scale(knn_features['Baths']) * baths_weight
    
    # Scale the target features using the same min and max as the dataset
    def scale_feature(value, feature_series, weight):
        min_val = feature_series.min()
        max_val = feature_series.max()
        if max_val - min_val == 0:
            scaled = 0.0
        else:
            scaled = (value - min_val) / (max_val - min_val)
        return scaled * weight
    
    target_scaled = np.array([
        scale_feature(given_size, knn_features['MyHome_Floor_Area_Value'], size_weight),
        scale_feature(given_beds, knn_features['Beds'], beds_weight),
        scale_feature(given_baths, knn_features['Baths'], baths_weight)
    ]).reshape(1, -1)
    
    # Initialize NearestNeighbors with Euclidean distance
    nbrs = NearestNeighbors(n_neighbors=len(knn_features_scaled), 
                            algorithm='auto', metric='euclidean')
    nbrs.fit(knn_features_scaled)
    
    # Find distances from the target to all properties
    knn_distances, _ = nbrs.kneighbors(target_scaled)
    
    # Assign the KNN distances to the dataframe
    df_filtered['KNN Distance'] = knn_distances.flatten()
    
    # Normalize KNN Distance using Min-Max Scaling
    df_filtered['KNN Distance_scaled'] = min_max_scale(df_filtered['KNN Distance'])
    
    # Combine the scaled distances using the specified weights
    df_filtered['Combined Score'] = (
        geo_weight * df_filtered['Geo Score'] + 
        (1 - geo_weight) * df_filtered['KNN Distance_scaled']
    )
    
    # Sort by the combined score in ascending order (lower is better)
    df_sorted = df_filtered.sort_values(by='Combined Score').reset_index(drop=True)
    
    return df_sorted

# --------------------------------------------
# 3. Execute Scoring
# --------------------------------------------

# Define the given property details
given_property = {
    'latitude': 53.279690,   # Example latitude (Blackrock DART station)
    'longitude': -6.181159,  # Example longitude (Blackrock DART station)
    'beds': 1.0,             # Number of beds
    'baths': 1.0,            # Number of baths
    'size': 100.0            # Size in square meters
}

# Define weights
geo_weight = 0.7        # 70% weight for Geo Distance
knn_weight = 0.3        # 30% weight for KNN
size_weight = 60.0      # 60% weight for size
beds_weight = 20.0      # 20% weight for beds
baths_weight = 20.0     # 20% weight for baths

# Maximum distance to consider (in kilometers)
max_distance_km = 3

# Decay rate for exponential decline
decay_rate = 1.0

# Calculate the combined score
df_sorted = calculate_combined_score(
    df_subset=df_subset,
    given_lat=given_property['latitude'],
    given_lon=given_property['longitude'],
    given_beds=given_property['beds'],
    given_baths=given_property['baths'],
    given_size=given_property['size'],
    geo_weight=geo_weight,
    size_weight=size_weight,
    beds_weight=beds_weight,
    baths_weight=baths_weight,
    max_distance_km=max_distance_km,
    decay_rate=decay_rate
)

# Check if df_sorted is not empty before proceeding
if not df_sorted.empty:
    # Display the top 20 rows of the sorted DataFrame
    print("Top 20 Properties Based on Combined Score:")
    display_columns = [
        'Address', 'Property Type', 'Beds', 'Baths', 
        'MyHome_Floor_Area_Value', 'Sold Asking Price', 
        'MyHome_Latitude', 'MyHome_Longitude', 
        'Geo Distance (km)', 'Geo Score', 'KNN Distance', 
        'KNN Distance_scaled', 'Combined Score'
    ]
    display(df_sorted[display_columns].head(20))
else:
    print("No properties match the criteria within the specified distance.")

```

```python
import pandas as pd
import numpy as np
import logging
import matplotlib.pyplot as plt
from IPython.display import display

# Configure logging for better traceability
logging.basicConfig(level=logging.INFO, format='%(levelname)s: %(message)s')

# --------------------------------------------
# 1. Data Preparation
# --------------------------------------------

def prepare_data(house_rows):
    """
    Prepare the dataset for analysis.
    """
    columns_subset_training = [
        'Address', 'Property Type', 'Beds', 'Baths', 'MyHome_Floor_Area_Value',
        'Sold Asking Price', 'MyHome_Latitude', 'MyHome_Longitude'
    ]
    
    df_subset = house_rows[columns_subset_training].copy()
    
    # Ensure 'Beds' and 'Baths' are treated as strings before extracting numeric values
    df_subset['Beds'] = df_subset['Beds'].astype(str).str.extract('(\d+)').astype(float)
    df_subset['Baths'] = df_subset['Baths'].astype(str).str.extract('(\d+)').astype(float)
    
    # Convert other relevant columns to numeric
    numeric_columns = ['MyHome_Floor_Area_Value', 'Sold Asking Price', 
                       'MyHome_Latitude', 'MyHome_Longitude']
    for col in numeric_columns:
        df_subset[col] = pd.to_numeric(df_subset[col], errors='coerce')
    
    # Drop rows with missing values in critical columns
    df_subset.dropna(subset=['Beds', 'Baths', 'MyHome_Floor_Area_Value', 
                             'MyHome_Latitude', 'MyHome_Longitude'], inplace=True)
    
    # Reset index after dropping rows
    df_subset.reset_index(drop=True, inplace=True)
    
    return df_subset

# --------------------------------------------
# 2. Helper Functions
# --------------------------------------------

def haversine_distance(lat1, lon1, lat2, lon2):
    """Calculate the Haversine distance between two points on the Earth."""
    R = 6371  # Earth radius in kilometers
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat / 2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2)**2
    c = 2 * np.arcsin(np.sqrt(a))
    return R * c

def min_max_scale(feature):
    """Perform Min-Max scaling on a feature."""
    min_val = feature.min()
    max_val = feature.max()
    return (feature - min_val) / (max_val - min_val) if max_val - min_val != 0 else feature - min_val

def exponential_decay(distance, max_distance_km, decay_rate=1.0):
    """Apply exponential decay to the distance."""
    return np.exp(-decay_rate * (distance / max_distance_km))

def visualize_properties(df_subset, given_property, max_distance_km):
    """Visualize property locations with respect to the target property."""
    plt.figure(figsize=(12, 10))
    plt.scatter(df_subset['MyHome_Longitude'], df_subset['MyHome_Latitude'], 
                alpha=0.3, label='All Properties')
    plt.scatter(given_property['longitude'], given_property['latitude'], 
                color='red', marker='*', s=300, label='Target Property')
    plt.xlabel('Longitude', fontsize=12)
    plt.ylabel('Latitude', fontsize=12)
    plt.title(f'Property Locations within {max_distance_km} km', fontsize=16)
    plt.legend(fontsize=10)
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# --------------------------------------------
# 3. Main Analysis Functions
# --------------------------------------------

def calculate_combined_score(df_subset, given_lat, given_lon, given_beds, 
                             given_baths, given_size, geo_weight=0.7, 
                             size_weight=60.0, beds_weight=20.0, 
                             baths_weight=20.0, max_distance_km=3, 
                             decay_rate=1.0):
    """Calculate a combined score based on geographic distance and feature similarity."""
    logging.info("Calculating geographic distances using Haversine formula.")
    
    df_subset['Geo Distance (km)'] = df_subset.apply(
        lambda row: haversine_distance(given_lat, given_lon, 
                                       row['MyHome_Latitude'], 
                                       row['MyHome_Longitude']), axis=1)
    
    logging.info(f"Distance Statistics:\n{df_subset['Geo Distance (km)'].describe()}")
    
    df_filtered = df_subset[df_subset['Geo Distance (km)'] <= max_distance_km].copy()
    
    if df_filtered.empty:
        logging.warning("No properties found within the specified geographic distance.")
        return df_filtered
    
    logging.info(f"Found {len(df_filtered)} properties within {max_distance_km} km.")
    
    df_filtered['Geo Score'] = exponential_decay(
        df_filtered['Geo Distance (km)'], 
        max_distance_km, 
        decay_rate
    )
    
    logging.info(f"Geo Score Statistics:\n{df_filtered['Geo Score'].describe()}")
    
    knn_features = df_filtered[['MyHome_Floor_Area_Value', 'Beds', 'Baths']].copy()
    
    knn_features_scaled = pd.DataFrame()
    knn_features_scaled['MyHome_Floor_Area_Value'] = min_max_scale(knn_features['MyHome_Floor_Area_Value']) * size_weight
    knn_features_scaled['Beds'] = min_max_scale(knn_features['Beds']) * beds_weight
    knn_features_scaled['Baths'] = min_max_scale(knn_features['Baths']) * baths_weight
    
    def scale_feature(value, feature_series, weight):
        min_val = feature_series.min()
        max_val = feature_series.max()
        scaled = (value - min_val) / (max_val - min_val) if max_val - min_val != 0 else 0.0
        return scaled * weight
    
    target_scaled = np.array([
        scale_feature(given_size, knn_features['MyHome_Floor_Area_Value'], size_weight),
        scale_feature(given_beds, knn_features['Beds'], beds_weight),
        scale_feature(given_baths, knn_features['Baths'], baths_weight)
    ]).reshape(1, -1)
    
    knn_distances = np.linalg.norm(knn_features_scaled.values - target_scaled, axis=1)
    
    df_filtered['KNN Distance'] = knn_distances
    df_filtered['KNN Distance_scaled'] = min_max_scale(df_filtered['KNN Distance'])
    
    df_filtered['Combined Score'] = (
        geo_weight * df_filtered['Geo Score'] + 
        (1 - geo_weight) * df_filtered['KNN Distance_scaled']
    )
    
    logging.info(f"Combined Score Statistics:\n{df_filtered['Combined Score'].describe()}")
    
    df_sorted = df_filtered.sort_values(by='Combined Score', ascending=False).reset_index(drop=True)
    
    logging.info("Combined scoring completed.")
    
    return df_sorted

def estimate_property_value_weighted_average(df_sorted, top_n=10):
    """Estimate the property value based on the weighted average of the top N similar properties."""
    top_properties = df_sorted.head(top_n).copy()
    
    if top_properties.empty:
        logging.warning("No properties available for estimation.")
        return None
    
    top_properties['Weight'] = 1 / (top_properties['Combined Score'] + 1e-5)
    top_properties['Weight'] /= top_properties['Weight'].sum()
    
    estimated_value = (top_properties['Sold Asking Price'] * top_properties['Weight']).sum()
    
    logging.info(f"Estimated Property Value (Weighted Average): €{estimated_value:,.2f}")
    return estimated_value

def estimate_property_value_median(df_sorted, top_n=10):
    """Estimate the property value based on the median price of the top N similar properties."""
    top_properties = df_sorted.head(top_n)
    if top_properties.empty:
        logging.warning("No properties available for median estimation.")
        return None
    estimated_value = top_properties['Sold Asking Price'].median()
    logging.info(f"Estimated Property Value (Median): €{estimated_value:,.2f}")
    return estimated_value

def summarize_combined_score_decay(df_sorted, num_bins=10):
    """
    Summarize the number of properties within different combined score intervals to show decay.
    
    Parameters:
    - df_sorted: DataFrame sorted by Combined Score in descending order.
    - num_bins: Number of bins to divide the combined scores into.
    
    Returns:
    - summary_df: DataFrame containing bin ranges and corresponding counts.
    """
    if df_sorted.empty:
        logging.warning("No data available to summarize combined score decay.")
        return pd.DataFrame()
    
    # Define bin edges based on combined scores
    bins = np.linspace(df_sorted['Combined Score'].min(), df_sorted['Combined Score'].max(), num_bins + 1)
    labels = [f"{round(bins[i], 3)} - {round(bins[i+1], 3)}" for i in range(num_bins)]
    
    df_sorted['Combined Score Bin'] = pd.cut(df_sorted['Combined Score'], bins=bins, labels=labels, include_lowest=True)
    
    summary = df_sorted['Combined Score Bin'].value_counts().sort_index()
    summary_df = summary.reset_index()
    summary_df.columns = ['Combined Score Range', 'Number of Properties (N)']
    
    logging.info("Combined Score Decay Summary:")
    logging.info(f"\n{summary_df}")
    
    return summary_df

# --------------------------------------------
# 4. Main Execution Function
# --------------------------------------------

def run_property_valuation_analysis(house_rows, given_property, geo_weight=0.7, 
                                    size_weight=60.0, beds_weight=20.0, 
                                    baths_weight=20.0, max_distance_km=3, top_n=10):
    """
    Main function to run the property valuation analysis.
    
    Parameters:
    - house_rows: DataFrame containing property data.
    - given_property: Dictionary with keys 'latitude', 'longitude', 'beds', 'baths', 'size'.
    - geo_weight: Weight for the geographic score in combined score calculation.
    - size_weight: Weight for the size feature.
    - beds_weight: Weight for the beds feature.
    - baths_weight: Weight for the baths feature.
    - max_distance_km: Maximum geographic distance to consider.
    - top_n: Number of top properties to include in valuation calculations.
    """
    # Prepare data
    df_subset = prepare_data(house_rows)
    
    # Calculate the combined score
    df_sorted = calculate_combined_score(
        df_subset=df_subset,
        given_lat=given_property['latitude'],
        given_lon=given_property['longitude'],
        given_beds=given_property['beds'],
        given_baths=given_property['baths'],
        given_size=given_property['size'],
        geo_weight=geo_weight,
        size_weight=size_weight,
        beds_weight=beds_weight,
        baths_weight=baths_weight,
        max_distance_km=max_distance_km,
        decay_rate=1.0
    )
    
    # Visualize properties
    # visualize_properties(df_subset, given_property, max_distance_km)
    
    if not df_sorted.empty:
        # Display the top 20 rows of the sorted DataFrame
        print("Top 20 Properties Based on Combined Score:")
        display_columns = [
            'Address', 'Property Type', 'Beds', 'Baths', 
            'MyHome_Floor_Area_Value', 'Sold Asking Price', 
            'MyHome_Latitude', 'MyHome_Longitude', 
            'Geo Distance (km)', 'KNN Distance', 
            'KNN Distance_scaled', 'Combined Score'
        ]
        display(df_sorted[display_columns].head(20))
        
        # Estimate property value using weighted average
        estimated_value = estimate_property_value_weighted_average(df_sorted, top_n=top_n)
        if estimated_value is not None:
            print(f"Estimated Property Value (Weighted Average) using top {top_n} properties: €{estimated_value:,.2f}")
        
        # Estimate property value using median
        estimated_value_median = estimate_property_value_median(df_sorted, top_n=top_n)
        if estimated_value_median is not None:
            print(f"Estimated Property Value (Median) using top {top_n} properties: €{estimated_value_median:,.2f}")
        
        # Summarize combined score decay
        decay_summary = summarize_combined_score_decay(df_sorted)
        if not decay_summary.empty:
            print("\nNumber of Properties within Combined Score Ranges (Decay):")
            display(decay_summary)
    else:
        print("No properties match the criteria within the specified distance.")
        # Suggest increasing the maximum distance
        user_input = input("Would you like to increase the maximum distance and retry? (yes/no): ").strip().lower()
        if user_input in ['yes', 'y']:
            try:
                new_max_distance_km = float(input("Enter the new maximum distance in kilometers: "))
                logging.info(f"Retrying with a new maximum distance of {new_max_distance_km} km.")
                run_property_valuation_analysis(
                    house_rows, given_property, geo_weight, 
                    size_weight, beds_weight, baths_weight, 
                    new_max_distance_km, top_n
                )
            except ValueError:
                logging.error("Invalid input for maximum distance. Please enter a numeric value.")
        else:
            print("Exiting without retrying.")

# --------------------------------------------
# 5. Usage Example
# --------------------------------------------

# Example usage:
# Assuming house_rows DataFrame is available
# house_rows = pd.read_csv('your_data.csv')

# Define the given property details
given_property = {
    'latitude': 53.279690,   # Example latitude (Blackrock DART station)
    'longitude': -6.181159,  # Example longitude (Blackrock DART station)
    'beds': 4.0,             # Number of beds
    'baths': 2.0,            # Number of baths
    'size': 200.0            # Size in square meters
}

# Example top_n parameter
top_n = 10  # You can change this value as needed

# Run the analysis
run_property_valuation_analysis(house_rows, given_property, top_n=top_n)
```

```python
# Import Necessary Libraries
import pandas as pd
import matplotlib.pyplot as plt

def analyze_top_n_prices(df, N):
    """
    Analyzes the top N rows of a DataFrame and returns the 25th percentile, median, and 75th percentile
    of the 'Sold Asking Price'. Also plots these statistics.

    Parameters:
    df (pd.DataFrame): The sorted DataFrame.
    N (int): Number of top rows to analyze.

    Returns:
    pd.DataFrame: A DataFrame with the statistics.
    """
    # Validate Input Parameters
    if not isinstance(df, pd.DataFrame):
        raise TypeError("The 'df' parameter must be a pandas DataFrame.")
    if 'Sold Asking Price' not in df.columns:
        raise ValueError("The DataFrame must contain a 'Sold Asking Price' column.")
    if not isinstance(N, int) or N <= 0:
        raise ValueError("The 'N' parameter must be a positive integer.")

    # Select the top N rows
    top_n = df.head(N)

    # Check if there are enough rows
    if top_n.empty:
        print(f"The DataFrame is empty. No data to analyze for top {N} properties.")
        return pd.DataFrame()

    # Calculate the percentiles
    percentiles = top_n['Sold Asking Price'].quantile([0.25, 0.5, 0.75]).rename(index={
        0.25: '25th Percentile',
        0.5: 'Median',
        0.75: '75th Percentile'
        
    })

    # Create a summary DataFrame
    summary_df = pd.DataFrame({
        'Statistic': ['25th Percentile', 'Median', '75th Percentile'],
        'Sold Asking Price (EUR)': [
            percentiles['25th Percentile'], 
            percentiles['Median'], 
            percentiles['75th Percentile']
        ]
    })

    # Print the summary
    print(f"Sold Asking Price Statistics for Top {N} Properties:")
    display(summary_df)

    return summary_df
    # Call the analyze_top_n_prices function
price_summary = analyze_top_n_prices(df_sorted, 20)
```
